# Task
Calculate the average of the 'G3' column in the `df` DataFrame, then identify and display all students whose 'G3' score is below this average, and finally summarize the characteristics of these below-average students.

## Calculate Average Grade

### Subtask:
Calculate the average score of the 'G3' column to determine the overall student performance.


**Reasoning**:
Calculate the average of the 'G3' column as per the instructions and store it in a variable. Then, print the calculated average.



```python
average_g3_score = df['G3'].mean()
print(f"The average G3 score is: {average_g3_score}")
```

```python
average_score = df['G3'].mean()
average_score
```

## Identify Below-Average Students and Above-Average Students

### Subtask:
Identify and display all students from the `df` DataFrame whose final grade ('G3') is below the calculated average.

#### Instructions
1. Filter the `df` DataFrame to select only those rows where the 'G3' column value is less than the `average_g3_score`.
2. Store this filtered DataFrame in a new variable, for example, `below_average_students`.
3. Display the first few rows of the `below_average_students` DataFrame to inspect the results.

**Reasoning**:
Now that the average 'G3' score is calculated, I will filter the original DataFrame to find students whose 'G3' score is below and above this average, as instructed in the subtask.



```python
above_average_students = df[df['G3'] > average_g3_score]
above_average_students[['G3']]
```

```python
below_average_students = df[df['G3'] < average_g3_score]
below_average_students[['G3']]
```

```python
above=(df['G3']> average_g3_score).sum()
print(above)
```

```python
below=(df['G3'] < average_g3_score).sum()
print(below)
```

## **Matplotlib**

--> used data visualization and Histogram

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df['G3'])
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df['G3'])
plt.xlabel('Final Grade')
plt.ylabel('Frequency')
plt.title('Distribution of Final Grades')
plt.show()
```

```python
#Barchart

#find the average marks for boy and girld
average_g3_score = df.groupby('sex')['G3'].mean()
average_g3_score

plt.figure()
average_g3_score.plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade')
```

```python
# scatter plot

plt.figure()
plt.scatter(df['studytime'], df['G3'])
plt.xlabel('studytime')
plt.ylabel('G3')
```

```python
plt.figure()
plt.scatter(df['studytime'], df['G2'])
plt.xlabel('studytime')
plt.ylabel('G2')
```

```python
#box plot

plt.figure()
plt.boxplot(df['G3'])
plt.title('Box Plot of final score')
plt.xlabel('G3')
plt.ylabel('G3')
```

```python
#line ploat
avg_marks=df[['G1','G2','G3']].mean()

plt.figure()
plt.plot(avg_marks.index, avg_marks.values,marker='o')
plt.xlabel('Subject')
plt.ylabel('Average Marks')
plt.title('Average Marks by Subject')
plt.show()
```

```python
passed =(df['G3']>=10).sum()
print("Number of student Passed",passed)
```

```python
#with barchart we find the number of student pass and fail

passed_students = df[df['G3'] >= 10].shape[0]
failed_students = df[df['G3'] < 10].shape[0]

plt.figure(figsize=(6, 4))
plt.bar(['Passed', 'Failed'], [passed_students, failed_students], color=['green', 'red'])
plt.xlabel('Result')
plt.ylabel('Number of Students')
plt.title('Number of Students: Pass vs. Fail (G3 >= 10)')
plt.show()
```

```python
#histogram absent distribution

plt.figure()
plt.hist(df['absences'], bins=10,color=('skyblue'))
plt.xlabel('Absences')
plt.ylabel('Frequency')
plt.title('Histogram of Absences')
plt.show()
```

```python
plt.figure()
plt.hist(df['absences'], bins=10, color='skyblue',edgecolor='black')
plt.xlabel('Absences')
plt.ylabel('Frequency')
plt.title('Histogram of Absences')
plt.box(True) # Add a border around the plot
plt.show()
```
