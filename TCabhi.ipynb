```python
# Re-define cat_cols and num_cols right before initializing the preprocessor to ensure they are up-to-date
cat_cols = X.select_dtypes(include='object').columns
num_cols = X.select_dtypes(exclude='object').columns

# building sklearn preprocessor
preprocessor=ColumnTransformer(transformers=[
    ('cat',OneHotEncoder(drop='first', handle_unknown='ignore'),cat_cols),
    ('num','passthrough',num_cols)
]
)
```

```python
# Apply the preprocessor to transform the training and testing features
X_train_transformed = preprocessor.fit_transform(x_train)
X_test_transformed = preprocessor.transform(x_test)

print("Shape of transformed training data:", X_train_transformed.shape)
print("Shape of transformed testing data:", X_test_transformed.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
# Create the machine learning pipeline
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(random_state=42))
])

print("Model pipeline created successfully.")
```

```python
for depth in[2, 4, 6, 10, None]:
  dt_pipeline = Pipeline(steps=[
      ('preprocessor', preprocessor),
      ('mode', DecisionTreeClassifier(max_depth=depth, random_state=42))
  ])

  dt_pipeline.fit(x_train, y_train)
  train_acc = dt_pipeline.score(x_train, y_train)
  y_pred = dt_pipeline.predict(x_test)
  test_acc = accuracy_score(y_test, y_pred)
  print(f"Depth={depth}, train ={train_acc: .2f}, Test={test_acc: .2f}")
```

```python
dt_model = dt_pipeline.named_steps['mode']
encoder = dt_pipeline.named_steps['preprocessor']

features_names = encoder.get_feature_names_out()

importances = pd.Series(
    dt_model.feature_importances_,
    index=features_names
).sort_values(ascending=False)

importances.head(10)
```
